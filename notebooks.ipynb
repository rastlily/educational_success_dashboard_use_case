```python
import pandas as pd

df = pd.read_csv("Statistik der allgemeinbildenden Schulen (Bundesländer, Geschlecht, 2013- 2022).csv", sep=";")
```

```python
df.head()
```

```python
df = df.ffill() # Replace NaN with the last available value
df
```

```python
# Renaming the first columns to have a consistent index after each column
df.rename(columns= {'männlich': 'männlich.0', 'weiblich':'weiblich.0', 'Insgesamt':'Insgesamt.0'}, inplace=True)
```

```python
df.head()
```

```python
new_df = pd.DataFrame(columns= ['Bundesland', 'Schulabschlüsse', 'Geschlecht', 'Jahr', 'Wert'])

year_list = [2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]

for index, row in df.iterrows():

    # Skipping the 'Total' rows
    if row.Schulabschlüsse == "Insgesamt":
        continue

    # For each column in the row
    # Skip the first two columns which are constant
    for column, value in row[2:].items():

        # Skip all "Insgesamt" columns
        if "Insgesamt" in column :
            continue
 
        # The index is the lenght of the data frame
        row_idx = new_df.shape[0]

        # Gender - without .1, .2, usw.
        gender = column.split('.')[0]

        # Select the year witht the index
        year = year_list[ int(column.split('.')[1]) ]

        # Add a new row with the values
        new_df.loc[row_idx] = [row.Bundesland, row.Schulabschlüsse, gender, year, value]
```

```python
new_df
```

```python
new_df.to_csv('fixed - Statistik der allgemeinbildenden Schulen.csv', sep=';', index=False)
```
